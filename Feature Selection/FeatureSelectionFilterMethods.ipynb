**Variance Threshold**

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel=VarianceThreshold(threshold=0.05)
sel.fit(X_train)
```

```python
sum(sel.get_support())
```

369 columns out of 562 in our training data have their variance over the set threshold

```python
columns=[]
for i in X_train:
    columns.append(i)
len(columns)
```

```python
X_train[columns[3]].corr(X_train[columns[14]])
```

```python
copy=X_train
ToBeDroppeed=[]
droppedCount=0
for i in range(561):
    for j in range(i+1,562):
        if((copy[columns[i]].corr(copy[columns[j]])>0.9) | (copy[columns[i]].corr(copy[columns[j]])< (-0.9))):
            if(columns[j] in ToBeDroppeed):
                pass
            else:
                ToBeDroppeed.append(columns[j])
                droppedCount+=1
len(ToBeDroppeed)        
```

```python
X_train.drop(columns=ToBeDroppeed,inplace=True)
X_train
```

**ANOVA for Feature Selection**

```python
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
```

```python
anova= SelectKBest(f_classif,k=100).fit(X_train,y_train)
### SelectKBest selects the 100 features that prform best in ANOVA test
columns=X_train.columns[anova.get_support()] #100 best features
```

```python
X_train=pd.DataFrame(X_train,columns=columns)
```
